```python
#  Calculate averages for the desired columns by Grade.
#in Purchse data analysis video
#use Pandas recap

Spending_bins= [0,585, 630, 645, 680]
Labels = [" <585", "$585-630", "$630-645", "$645-680"] 
spendingScoresDF= per_school_summary.copy()
spendingScoresDF["Spending Ranges(Per Student)"]=pd.cut(perschoolCapita, Spending_bins, labels=Labels)
spendingScoresDF


```

```python
spendingmathDF = spendingScoresDF.groupby(["Spending Ranges(Per Student)"])["Average Math Score"].mean()
spendingreadingDF = spendingScoresDF.groupby(["Spending Ranges(Per Student)"])["Average Reading Score"].mean()
spendingpassingmathSeries = spendingScoresDF.groupby(["Spending Ranges(Per Student)"])["% Passing Math"].mean()
spendingpassingreadingSeries = spendingScoresDF.groupby(["Spending Ranges(Per Student)"])["% Passing Reading"].mean()
overallpassingspendingSeries = spendingScoresDF.groupby(["Spending Ranges(Per Student)"])["Overall Passing Rate"].mean()

```

```python
spendingSummaryDF=pd.DataFrame(
    {
        "Average Math Scores":spendingmathDF,
        "Average Reading Scores":spendingreadingDF,
        "% Passing Math":spendingpassingmathSeries,
        "% Passing Reading":spendingpassingreadingSeries,
        "% Overall Passing":overallpassingspendingSeries
       
    }
)
spendingSummaryDF
```

```python
#Use the following code to bin the per_school_summary.
#size_bins = [0, 1000, 2000, 5000]
#labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
#Use pd.cut on the "Total Students" column of the per_school_summary DataFrame.
#Create a DataFrame called size_summary that breaks down school performance based on school size (small, medium, or large).
#Scores by School Type
#Use the per_school_summary DataFrame from the previous step to create a new DataFrame called type_summary.
#This new DataFrame should show school performance based on the "School Type".

size_bins=[0,1000,2000,5000]
labels2 = ["Small (<1000)","Medium (1000-2000)", "Large (2000-5000)"]
spendingScoresDF["School Size"]= pd.cut(spendingScoresDF["Total Students"], size_bins, labels=labels2)
spendingScoresDF
```

```python
#  Calculate averages for the desired columns.

spending_math_score = spendingScoresDF.groupby(["School Size"])["Average Math Score"].mean()
spending_reading_score = spendingScoresDF.groupby(["School Size"])["Average Reading Score"].mean()
spending_passing_math = spendingScoresDF.groupby(["School Size"])["% Passing Math"].mean()
spending_passing_reading = spendingScoresDF.groupby(["School Size"])["% Passing Reading"].mean()
overall_passing_spending = spendingScoresDF.groupby(["School Size"])["Overall Passing Rate"].mean()

```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
        "Average Math Score": spending_math_score,
        "Average Reading Score": spending_reading_score,
        "% Passing Math": spending_passing_math,
        "% Passing Reading": spending_passing_reading,
        "% Overall Passing": overall_passing_spending
})

# Display results
size_summary
```

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = spendingScoresDF.groupby(["School Type"])["Average Math Score"].mean()
average_reading_score_by_type = spendingScoresDF.groupby(["School Type"])["Average Reading Score"].mean()
average_percent_passing_math_by_type = spendingScoresDF.groupby(["School Type"])["% Passing Math"].mean()
average_percent_passing_reading_by_type = spendingScoresDF.groupby(["School Type"])["% Passing Reading"].mean()
average_percent_overall_passing_by_type = spendingScoresDF.groupby(["School Type"])["Overall Passing Rate"].mean()
```

```python
# Assemble the new data by type into a DataFrame called `type_summary`
type_summary = pd.DataFrame({
        "Average Math Score": average_math_score_by_type,
        "Average Reading Score": average_reading_score_by_type,
        "% Passing Math": average_percent_passing_math_by_type,
        "% Passing Reading": average_percent_passing_reading_by_type,
        "% Overall Passing": average_percent_overall_passing_by_type
})

# Display results
type_summary
```
